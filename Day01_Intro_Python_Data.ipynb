```python
# Variable et types
age = 10
pi = 3.14
name = "Mayssam"
print(age, pi, name)
print(type(age), type(pi), type(name))

# Liste et boucle
fruits = ["pomme", "bananne", "orange"]
for f in fruits:
  print("jaime", f)

# Fonction simple
def moyenne(l):
    return sum(l)/len(l)

temps = [12,15,11,14]
print("Moyenne:", moyenne(temps))



```

```python
import pandas as pd

#creons un petit dataset
data = {
    "user_id": [1,2,3,4,5],
    "age": [23, 37, 31, 45, 29],
    "visits_last_month": [2, 5, 1, 7, 3]
}
df = pd.DataFrame(data)
print(df)

#premiere statistique
print("Age moyen ; ", df["age"].mean())
print("Age median :", df["age"].median())
print("Age min :", df["age"].min())
print("Age max :", df["age"].max())
print("Nombre moyen de visites ; ", df["visits_last_month"].mean())
print("\nResume :\n", df.describe())

# Add column
df["is_active"] = df["visits_last_month"] >= 3
print(df)

# Add age_10ans column
df["age_10ans"] = df["age"] + 10

# Filter for active users
active_users_df = df[df["is_active"]]

# Calculate mean of age_10ans for active users
mean_age_active_users = active_users_df["age_10ans"].mean()

print("\nMoyenne de age_10ans pour les utilisateurs actifs :", mean_age_active_users)

df.groupby("is_active").agg({
  "user_id":"count",
  "visits_last_month":"mean",
  "age" : "mean"
})
```
